## 3. *На языке Python предложить алгоритм вычисляющий численно предел с точностью $\epsilon = 10^{-7}$

```python
from math import factorial


def func(n):
    return n/(factorial(n)**(1/n))


e = 10**(-7)
i = 1
while abs(func(i+1)-func(i)) > e:
    i += 1
print(f'Предел равен {func(i+1)}')
print(f'Достаточно вычислить {i+1}-й член последовательности')
```
